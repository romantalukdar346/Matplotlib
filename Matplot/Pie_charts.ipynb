```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# New thing that i learn
### 1.wedgeprops={'edgecolor':'color name'}
### 2.startangle
### 3.shadow
### 4.explode

```python
plt.style.use("fivethirtyeight")
slices=[60,40,30,20]
labels=['Sixty','Fourty','Extra1','Extra2']
colors=['#008fd5','#fc4f30','#e5ae37','#6d904f']
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python


slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode=[0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode=[0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,shadow=True,wedgeprops={'edgecolor':'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode=[0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,shadow=True,wedgeprops={'edgecolor':'black'},startangle=90)
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode=[0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,shadow=True,wedgeprops={'edgecolor':'black'}
        ,startangle=90,autopct='%1.1f%%')
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
