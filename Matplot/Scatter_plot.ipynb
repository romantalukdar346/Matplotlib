```python
plt.style.use('seaborn')

plt.scatter(x,y,s=100)

plt.tight_layout()

plt.show()
```

```python
plt.scatter(x,y,s=100,color='green')

plt.tight_layout()

plt.show()
```

```python
plt.scatter(x,y,s=100,color='green',edgecolor='black',linewidth=1,alpha=.75)

plt.tight_layout()

plt.show()
```

```python

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x,y,s=100, c= colors,cmap='Greens' ,edgecolor='black',linewidth=1,alpha=.75)

plt.tight_layout()

plt.show()
```

```python
plt.scatter(x,y,s=100, c= colors,cmap='Greens' ,edgecolor='black',linewidth=1,alpha=.75)

cbar= plt.colorbar()
cbar.set_label('Satisfaction')
plt.tight_layout()

plt.show()
```

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x,y,s=sizes, c= colors,cmap='Greens' ,edgecolor='black',linewidth=1,alpha=.75)

cbar= plt.colorbar()
cbar.set_label('Satisfaction')
plt.tight_layout()

plt.show()
```

```python
str= r'C:\Users\KIIT\OneDrive\Desktop\code 1\Machine_Learning\Scatter_plot_yt.csv'
data=pd.read_csv(str)
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.scatter(view_count,likes,edgecolor='black',linewidth=1,alpha=.75)


plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
```

```python
plt.scatter(view_count,likes,edgecolor='black',linewidth=1,alpha=.75)

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
```

```python
plt.scatter(view_count,likes,edgecolor='black',c=ratio,cmap='summer',linewidth=1,alpha=.75)

plt.xscale('log')
plt.yscale('log')

cbar=plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
```
